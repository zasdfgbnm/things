```python
import matplotlib.pyplot as plt
sort_labels = ['int64_t', 'DynamicType1', 'DynamicType2', 'DynamicType3',
               'DynamicType4', 'DynamicType5', 'DynamicType6', 'DynamicType7',
               'DynamicType8', 'DynamicType9', 'DynamicType10']
sort_times = [55.1, 69.6, 92.6, 94.8, 99.7, 102, 131, 128, 132, 133, 136]

colors = ['grey'] + ['#76b900'] * (len(sort_labels) - 1)
plt.figure(figsize=(10,6))  # Increase the size for better visualization
bars = plt.bar(sort_labels, sort_times, color=colors)
plt.ylabel('Time (ms) - smaller is better')
plt.title('Sort Benchmark')
plt.xticks(rotation=45)    # Rotate labels for better visualization
# Annotate each bar with its height (the time value)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.savefig("sort.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

knn_labels = ['Native struct', 'Dictionary']
knn_times = [0.023, 2.95]

bars = plt.bar(knn_labels, knn_times, color=['grey', '#76b900'])
plt.yscale('log')  # Set the y-axis to a logarithmic scale
plt.ylabel('Time (ms) - Log Scale, smaller is better')
plt.ylim(0.01, 10)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, round(yval, 3), ha='center', va='bottom')


plt.title('kNN Benchmark')
plt.savefig("benchmark-knn.png")
plt.show()
```
